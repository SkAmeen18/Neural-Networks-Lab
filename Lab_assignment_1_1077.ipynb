```python
#Task 1️: Dataset Creation
# Features: [Study Hours, Attendance percentage]
features = [
    [5,80],[2,40],[8,90],[1,30],[7,75],[4,70],[3,50],[9,95],[2,85],[6,60],[1,80],[5,55],[6,65]
]

# Labels: 1 = Pass, 0 = Fail
#A student passes if their study hours are 4 or more AND their attendance is 70% or more
labels = [1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0, 0, 0]

print(f"Dataset initialized with {len(features)} samples.")
```

```python
#Task 2️: Perceptron Initialization

weights = [0.25, 0.35]
bias = 1
learning_rate = 0.01
epochs = 50
```

```python
#Task 3️: Activation Function

def activation_function(inputs, weights, bias):
    # Step A: Calculate weighted sum
    # Formula: (x1 * w1) + (x2 * w2) + bias
    weighted_sum = bias
    for i in range(len(inputs)):
        weighted_sum += inputs[i] * weights[i]

    # Step B: Threshold (Step Function)
    if weighted_sum >= 0:
        return 1
    else:
        return 0
```

```python
#Task 4️: Training Loop

print("Training in progress...")

for epoch in range(epochs):
    total_error = 0
    for i in range(len(features)):
        # 1. Predict
        prediction = activation_function(features[i], weights, bias)

        # 2. Calculate Error (Target - Prediction)
        error = labels[i] - prediction
        total_error += abs(error)

        # 3. Update Weights and Bias
        if error != 0:
            for j in range(len(weights)):
                weights[j] += learning_rate * error * features[i][j]
            bias += learning_rate * error

    # Bonus: View progress
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch + 1}: Errors = {total_error}")

print("Training finished.")
```

```python
#Task 5️: User Input Testing

print("\n--- Test the Trained Model ---")
in_study = float(input("Enter Study Hours per day: "))
in_attend = float(input("Enter Attendance Percentage: "))

test_input = [in_study, in_attend]
final_prediction = activation_function(test_input, weights, bias)

if final_prediction == 1:
    print("Result: This student is likely to PASS.")
else:
    print("Result: This student is likely to FAIL.")
```

Summary

Dataset: I randomly made date of 13 students. I have used a threshold (Study > 4 and Attendance > 70) to ensure the data was Linearly Separable.

Learning Rate: I used 0.01. 0.1 will be too aggressive;the attendance values go up to 100, a high learning rate can cause the weights to jump to massive numbers instantly.

Verification: n the first 10 epochs, the error was high (around 8), meaning the random initial weights were guessing incorrectly for most students.the error stays at 8 and never decreases, it means the Perceptron is unable to find a line that separates your Pass/Fail data. This usually happens if the data is non-linearly separable.
